```python
df = pd.read_csv(r'C:\Users\noahm\Downloads\DATA\heart_2020_cleaned.csv')
```

```python
df.drop(['AgeCategory', 'GenHealth', 'Race'], axis=1,inplace=True)
```

```python
df[df.columns[0:]] = df[df.columns[0:]].replace({'No':0, 'Yes': 1, 'Male':0, 'Female':1, 'No, borderline diabetes':0, 'Yes, borderline diabetes':1, 'Yes (during pregnancy)':0}).astype('int64')
```

```python
df.corr(method = 'pearson')
```

```python
matrix=df.corr().round(2)
print(matrix)
```

```python
matrix = df.corr().round(2)
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(matrix, annot=True)
plt.show()
```
